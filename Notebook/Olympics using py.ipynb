# **OLYMPICS DATA ANALYSIS** 

```python
#@title Importing dataset

url1 = 'https://raw.githubusercontent.com/tejaschaudhari192/Olympics-data-analysis/master/Dataset/athlete_events.csv'
url2 = 'https://raw.githubusercontent.com/tejaschaudhari192/Olympics-data-analysis/master/Dataset/noc_regions.csv'
```

```python
#@title Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# athletes = pd.read_csv('C:/Users/Tejas/Teja Server ⚡/programming/Projects/Olympics/Dataset/athlete_events.csv')
athletes = pd.read_csv(url1)
# regions = pd.read_csv('C:/Users/Tejas/Teja Server ⚡/programming/Projects/Olympics/Dataset/noc_regions.csv')
regions = pd.read_csv(url2)
```

```python
athletes.head()
```

```python
regions.head()
```

```python
#Join The Dataframes
athletes_df = athletes.merge(regions, how = 'left', on = 'NOC')
athletes_df.head() 
```

```python
athletes_df.shape
```

```python
#Column Name Consistent
athletes_df.rename( columns = {'region' : 'Regions', 'notes' : 'Notes'}, inplace= True)
```

```python
#Top Coutries Participation

top10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top10_countries
```

```python
#Plot For Top 10 Coutries
plt.figure(figsize = (10,5) )
plt.title('Overall Participation by Country')
sns.barplot(x=top10_countries.index, y = top10_countries, palette = 'Set2')
```

```python
#Age Distribution of The Participation
plt.figure(figsize = (10,5) )
plt.title('Age Distributions')
plt.xlabel('Age')
plt.ylabel('Numbr of participants')
plt.hist(athletes_df.Age, bins = np.arange(10,80,2), color = 'red', edgecolor = 'yellow')
```

```python
#Winter Olympics Sports
winter_sports = athletes_df[athletes_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
#Summer Olympics Sports
summer_sports = athletes_df[athletes_df.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
#Male and Female Participants
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
# Pie Plot For Male and Female Athletes
plt.figure(figsize = (10,5))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels=gender_counts.index, autopct = '%1.1f%%', shadow = True)
```

```python
#Total Medals
athletes_df.Medal.value_counts()
```

```python
#Total Number of Female Athletes in Each Olympics.
female_participants = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')][['Sex','Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.head()
```

```python
womenOlympics = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x='Year', data=womenOlympics, palette ="Spectral")
plt.title('Women Participation')
```

```python
part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:, 'F'].plot()
plt.title('Plot of Female Athletes Over Time')
```

```python
#Gold Medal Athletes

goldmedals = athletes_df[(athletes_df.Medal == 'Gold')]
```

```python
#Take Only the Values That Are Different Feom NaN
goldmedals = goldmedals[np.isfinite(goldmedals['Age'])]
```

```python
#Gold Beyond 60

goldmedals['ID'][goldmedals['Age'] > 60]. count()
```

```python
sporting_event = goldmedals['Sport'][goldmedals['Age']>60]
sporting_event
```

```python
#Plot For Sporting Event

plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold Medals for Athletes Over 60 Years')
```

```python
#Gold Medal From Each Country

goldmedals.Regions.value_counts().reset_index(name='Medal').head(5)
```

```python
totalGoldMedals = goldmedals.Regions.value_counts().reset_index(name='Medal').head(6)
g = sns.catplot(x="index", y="Medal", data=totalGoldMedals, height=5,kind="bar", palette="rocket")
g.despine(left=True)
g.set_xlabels("Top 5 Countries")
g.set_ylabels("Number of Medals")
plt.title('Gold Medal Per Country')
```

```python
#Rio Olympics

max_year = athletes_df.Year.max()
print(max_year)

team_names = athletes_df[(athletes_df.Year == max_year) & (athletes_df.Medal == 'Gold')].Team

team_names.value_counts().head(10)
```

```python
sns.barplot(x=team_names.value_counts().head(20), y=team_names.value_counts().head(20).index)

plt.ylabel(None);
plt.xlabel('Country Wise Medal For The Year 2016');
```

```python
not_null_medals = athletes_df[(athletes_df['Height'].notnull()) & (athletes_df['Weight'])]
```

```python
plt.figure(figsize = (12, 10))
axis = sns.scatterplot(x="Height", y="Weight", data=not_null_medals, hue="Sex")
plt.title('Height vs Weight of Olympics Medalists')
```
